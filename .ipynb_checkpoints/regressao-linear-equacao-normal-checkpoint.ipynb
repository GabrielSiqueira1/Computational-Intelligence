<div align="justify">
    <p>Utilizando o arquivo data2.txt, implementaremos uma regressão linear em três dimensões através da equação normal.</p>
</div>

```python
import random
import math
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

with open("data2.txt", "r", encoding="utf-8") as file:
    coordinates = file.readlines()

# Separação das coordenadas entre os vetores x e y
x1 = []
x2 = []
y = []
for line in coordinates:
    coord = line.split(",")
    x1.append(float(coord[0]))
    x2.append(float(coord[1]))
    y.append(float(coord[2]))
```

```python
# Aplicando a equação normal
x0 = np.ones(len(x1))
matriz_x = np.transpose(np.array([x0,x1,x2]))
resultado = (np.linalg.inv(np.transpose(matriz_x) @ matriz_x) @ np.transpose(matriz_x)) @ y
theta_0 = resultado[0]
theta_1 = resultado[1]
theta_2 = resultado[2]

# Exibindo os resultados
r, s = np.meshgrid(x1, x2)
h = theta_0 + theta_1*r + theta_2*s
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(r, s, h, 
                cmap=cm.coolwarm,
                alpha=1,        
                linewidth=0,
                antialiased=False)

ax.scatter(x1, x2, y, zdir='z', s=200, color='black')
fig.colorbar(surf, shrink=0.7, aspect=10)
ax.set_xlabel('$X$',rotation=150)
ax.set_ylabel('$Y$')
ax.set_zlabel('$Z$', fontsize=10)
```
