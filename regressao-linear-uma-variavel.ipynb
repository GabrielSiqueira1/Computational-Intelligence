```python
import random
import matplotlib.pyplot as plt
import numpy as np

with open("data1.txt", "r", encoding="utf-8") as file:
    coordinates = file.readlines()

# Separação das coordenadas entre os vetores x e y
x = []
y = []
for line in coordinates:
    coord = line.split(",")
    x.append(float(coord[0]))
    y.append(float(coord[1]))
```

```python
# Estabelecimento da hipótese

theta_0 = random.random()
theta_1 = random.random()
m = len(x)
epoca = []
alpha = 0.01
it = 0
LIMITE = 10000
```

```python
plt.plot(x,y,'ro')
plt.title('Amostra dos dados do arquivo 1')
plt.xlabel('População da cidade em 10000')
plt.ylabel('Lucro em R$10000')
plt.show()
```

```python
# Estabelecimento do código principal

while it < LIMITE:

    h = []
    for i in range(0, len(x)):
        h.append(theta_0+theta_1*x[i])

    erro = 0
    for i in range(0, len(h)):
        erro += (h[i]-y[i])**2
    J = 1/(2*m)*erro

    derivada_theta_0 = 0
    derivada_theta_1 = 0
    for j in range(0, len(h)):
        derivada_theta_0 += (h[j]-y[j])
        derivada_theta_1 += (h[j]-y[j])*x[j]
    derivada_theta_0 = 1/m*derivada_theta_0
    derivada_theta_1 = 1/m*derivada_theta_1

    theta_0 -= alpha*derivada_theta_0
    theta_1 -= alpha*derivada_theta_1
    
    epoca.append(erro) #SSE
    it += 1

h = []
for i in range(0, len(x)):
    h.append(theta_0+theta_1*x[i])
plt.plot(x,y,'ro')
plt.plot(x,h)
plt.title('Processo de regressão linear')
plt.xlabel('População da cidade em 10000')
plt.ylabel('Lucro em R$10000')
plt.show()
```

```python
ts = np.linspace(0, LIMITE, LIMITE)

print(f"Último valor de erro: {epoca[-1]}")
plt.plot(ts,epoca)
plt.title('Redução do erro')
plt.xlabel('Iterações')
plt.ylabel('Variação do erro')
plt.show()
```
